# Creating Dataset

```python
def create_dataset(img_folder,size):
    # size=(900,600)
    img_data_array=[]
    class_name=[]
    for dir1 in os.listdir(os.path.join(img_folder)):
        for file in os.listdir(os.path.join(img_folder, dir1)):
            image_path= os.path.join(img_folder, dir1,  file)

            image= cv2.imread( image_path,cv2.IMREAD_GRAYSCALE)#IMREAD_GRAYSCALE)# cv2.COLOR_BGR2RGB)
            image=cv2.resize(image, size,interpolation = cv2.INTER_AREA)
            image=np.array(image)
            image = image.astype('float32')
            image /= 255 
            img_data_array.append(image)
            class_name.append(dir1)

    target_dict={k: v for v, k in enumerate(np.unique(class_name))}
    target = [target_dict[class_name[i]] for i in range(len(class_name))]
    target=np.asarray(target)
    
    
    img_data_array=np.asarray(img_data_array, dtype=np.float32)
    
    return img_data_array,target
        
```

```python
size=(64,64)
train_path=r"D:\Jupyter Files\My freelancing project\FACEBOOK\Disease Classifier\Splitted_Data\train"
test_path=r"D:\Jupyter Files\My freelancing project\FACEBOOK\Disease Classifier\Splitted_Data\test"
val_path=r"D:\Jupyter Files\My freelancing project\FACEBOOK\Disease Classifier\Splitted_Data\val"
#this converts images to numpy arrays
X_train,Y_train=create_dataset(train_path,size)
X_test,Y_test=create_dataset(test_path,size)
X_validation,Y_validation=create_dataset(val_path,size)
```
